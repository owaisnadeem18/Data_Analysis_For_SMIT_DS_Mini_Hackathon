# Assignment: Find the Maximum Value of each and every column in our data set . Also mention the index number on which that specific data set lies:

```python
# code: (Solution of Assignment)

maximum_value_of_every_column = data.describe().loc["max"]
indx_of_maximum_value = data.idxmax()

result = pd.DataFrame({
    "Maximum Value": maximum_value_of_every_column ,
    '''Index No.''': indx_of_maximum_value
})

result
```
